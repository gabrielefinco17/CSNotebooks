## Creating a Series
A Series is a one-dimensional array-like object containing a sequence of values and an associated array of data labels, called its index.

```python
s = pd.Series([10, 20, 30, 40])
print(s)
```

```python
t = pd.Series([130, 120, 160, 140], index=['a', 'b', 'c', 'd'])
print(t)
```

```python
print("Original Series s:")
print(s)

print("\nOriginal Series t:")
print(t)

# 1. Scalar arithmetic operations
print("\nScalar addition (s + 5):")
print(s + 5)

print("\nScalar multiplication (t * 2):")
print(t * 2)

# 2. Universal functions (ufuncs)
# Pandas Series integrate seamlessly with NumPy ufuncs
print("\nApplying np.sqrt to s:")
print(np.sqrt(s))

print("\nApplying np.log to t:")
print(np.log(t))

# 3. Element-wise operations between two Series
# (Note: This will align by index. If indices don't match, it will result in NaN)
print("\nElement-wise addition (s + s):")
print(s + s)

# Create a new series for demonstration of index alignment
s_subset = pd.Series([1, 2], index=[0, 1])
print("\nSubset of s:")
print(s_subset)
print("\nElement-wise addition (s + s_subset) - demonstrating index alignment:")
print(s + s_subset)

# 4. Comparison operations
print("\nComparison operation (s > 25):")
print(s > 25)

print("\nComparison operation (t == 130):")
print(t == 130)
```

```python
# I can create a dict form a serie...
f = t.to_dict()
print(f)

# ....and viceversa
d = pd.Series(f)
print(d)
```

## Creating a DataFrame
A DataFrame is a two-dimensional, size-mutable, and heterogeneous data structure with named columns each of which can be a different value type. It can be defined as a **dictionary of Series**

```python
data = {
    'Name': ['Anna', 'Luca', 'Marco', 'Julia','Toni','Bepi', 'Nane'],
    'Age': [17, 18, 16, 17, 18, 19, 18],
    'Class': ['5A', '5B', '4A', '5A', '5C','5D', '5A']
}

df = pd.DataFrame(data)
df
```

```python
# Retrieve the 'Age' column as a Series
age_series = df['Age']
print(age_series)
print(type(age_series))
```

```python
print("First 5 rows of the DataFrame:")
print(df.head())

print("\nLast 3 rows of the DataFrame:")
print(df.tail(3))
```

* count: The number of non-null values.
* mean: The average value.
* std: The standard deviation.
* min: The minimum value.
* 25%: The first quartile (25th percentile).
* 50%: The median (50th percentile).
* 75%: The third quartile (75th percentile).
* max: The maximum value.

## Loading Data from CSV Files

Pandas provides the `read_csv()` function to easily load data from a CSV (Comma Separated Values) file into a DataFrame. This is one of the most common ways to get external data into your pandas workflow.

### Key parameters of `read_csv()`:
*   **filepath_or_buffer**: The path to the CSV file.
*   **sep**: Delimiter to use. Defaults to `','`.
*   **header**: Row number to use as the column names, and a zero-based index. Defaults to `0`.
*   **index_col**: Column(s) to use as the row labels of the DataFrame. Defaults to `None`.
*   **dtype**: Dictionary of columns and their data types.
*   **parse_dates**: List of column names to parse as dates.

Let's create a dummy CSV file and then load it into a DataFrame.

```python
import pandas as pd

# Create a dummy CSV from a db export

# Load the CSV file into a DataFrame
df_csv = pd.read_csv('sample_data.csv')

print("DataFrame loaded from 'sample_data.csv':")


# Then load directly from a URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df_titanic = pd.read_csv(url)
print("\nDataFrame loaded from URL (Titanic dataset head):")
display(df_titanic.head())
```

### Counting Unique Values

```python
# Count the number of passengers in each class
print("Count of passengers per Pclass:")
display(df_titanic['Pclass'].value_counts())

# Count the number of male and female passengers
print("\nCount of passengers by Sex:")
display(df_titanic['Sex'].value_counts())

# Count the number of survivors vs. non-survivors
print("\nSurvival counts (0 = No, 1 = Yes):")
display(df_titanic['Survived'].value_counts())
```

### Calculating Averages

```python
# Calculate the average age of passengers
print(f"Average age of all passengers: {df_titanic['Age'].mean():.2f}")

# Calculate the average fare paid by passengers
print(f"Average fare paid by all passengers: {df_titanic['Fare'].mean():.2f}")

# Calculate the average age for survivors vs. non-survivors
print("\nAverage age by survival status:")
display(df_titanic.groupby('Survived')['Age'].mean())

# Calculate the average survival rate for each gender
print("\nSurvival rate by Sex:")
display(df_titanic.groupby('Sex')['Survived'].mean())
```

### Filtering Data

```python
# Filter for passengers who survived
print("Survivors:")
display(df_titanic[df_titanic['Survived'] == 1].head())

# Filter for female passengers older than 30
print("\nFemale passengers older than 30:")
display(df_titanic[(df_titanic['Sex'] == 'female') & (df_titanic['Age'] > 30)].head())
```

### Sorting Data

```python
# Sort passengers by age in ascending order
print("Passengers sorted by Age (ascending):")
display(df_titanic.sort_values(by='Age').head())

# Sort passengers by fare in descending order, then by age in ascending order
print("\nPassengers sorted by Fare (descending) and then Age (ascending):")
display(df_titanic.sort_values(by=['Fare', 'Age'], ascending=[False, True]).head())
```
